```python
class Visualization:
    def __init__(self, df):
        self.df = df
        
    def bar_plot(self, ax, col):
        df_col = self.df[col].value_counts()
        palette = sns.color_palette("tab10", len(df_col))
        colors = [palette[i] for i in range(len(df_col))]
        
        ax.bar(df_col.index, df_col.values, color = colors)
        ax.set_title(f"{col} - Frequency")
        ax.set_xlabel(f"{col}")
        ax.set_ylabel(f"Frequency")
        
    def hist_plot(self, ax, col):
        palette = sns.color_palette("tab10")
        ax.hist(self.df[col], color = palette[2], bins = 40)
        ax.set_title(f"{col} - Frequency")
        ax.set_xlabel(f"{col}")
        ax.set_ylabel(f"Frequency")
    
    def pie_chart(self, ax, col):
        df_col = self.df[col].value_counts()
        palette = sns.color_palette("tab10", len(df_col))
        ax.pie(df_col.values, labels=df_col.index, autopct='%1.1f%%', colors=palette)
        ax.set_title(f"{col} Distribution")
        
    def box_plot(self, ax, col):
        palette = sns.color_palette("tab10")
        sns.boxplot(data=self.df, x=col, ax=ax, palette=palette)
        ax.set_title(f"{col} Box Plot")
        
    def combined_plots(self, cols, plt_fn, fig_size):
        n_col = 2
        n_row = (len(cols) + n_col - 1)//n_col
        fig, axes = plt.subplots(n_row, n_col,figsize=fig_size)
        
        axes = axes.flatten()
        
        for i, col in enumerate(cols):
            plt_fn(axes[i], col)
            
        for j in range(i + 1, len(axes)):
            fig.delaxes(axes[j])
            
        plt.tight_layout()
        plt.show()
```

### Feature Engineering

```python
class FeatureEngineering:
    def __init__(self, df):
        self.df = df
    
    def split_data(self):
        self.df[['PassengerGroup','PassengerSeat']] = self.df['PassengerId'].str.split("_", expand=True)
        self.df[['CabinDeck','CabinNum','CabinSide']] = self.df['Cabin'].str.split("/", expand=True)
        return self.df
    
    def null_handling(self):
        self.df["VRDeck"] = self.df["VRDeck"].fillna(self.df["VRDeck"].median())
        self.df["Spa"] = self.df["Spa"].fillna(self.df["Spa"].median())
        self.df["ShoppingMall"] = self.df["ShoppingMall"].fillna(self.df["ShoppingMall"].median())
        self.df["FoodCourt"] = self.df["FoodCourt"].fillna(self.df["FoodCourt"].median())
        self.df["RoomService"] = self.df["RoomService"].fillna(self.df["RoomService"].median())
        
        for i in self.df[self.df["CryoSleep"].isnull()].index:
            if self.df["RoomService"].iloc[i] + self.df["FoodCourt"].iloc[i] + self.df["ShoppingMall"].iloc[i] + self.df["Spa"].iloc[i] + self.df["VRDeck"].iloc[i]==0:
                self.df["CryoSleep"].iloc[i] = True
            else:
                self.df["CryoSleep"].iloc[i] = False

                
        return self.df
    
    def drop_cols(self, cols):
        return self.df.drop(cols, axis = 1)
    
    def new_feat(self, data):
        data["infant"] = data["Age"] <= 1
        data["kid"] = (data["Age"] <= 9) & (data["Age"] > 1)
        data["child"] = (data["Age"] <= 18) & (data["Age"] > 9)
        data["yound"] = (data["Age"] <= 30) & (data["Age"] > 18)
        data["adult"] = (data["Age"] <= 50) & (data["Age"] > 30)
        data["old"] = data["Age"] > 50
        
        data['family_name'] = data['Name'].str.split(' ', n=1, expand=True)[1]
        
        return data
```

```python
def set_age_based_on_vip(row):
    if pd.isna(row['Age']):
        if row['VIP'] == False:         
            row['Age'] = np.random.choice(range(60, 75)) if np.random.rand() > 0.5 else np.random.choice(range(1, 25))
        elif row['VIP'] == True:
            row['Age'] = np.random.randint(26, 59)
    return row


def set_vip_based_on_age(row):
    if pd.isna(row['VIP']):
        if row['Age'] < 25 or row['Age'] > 60:
            row['VIP'] = False
        else:
            row['VIP'] = True
    return row

train_data = train_data.apply(set_age_based_on_vip, axis=1)
train_data = train_data.apply(set_vip_based_on_age, axis=1)
```

```python
def populate_homeplanet(row):
    if pd.isna(row['HomePlanet']):
        if row['VIP'] == False:
            row['HomePlanet'] = 'Earth'
        elif row['VIP'] == True:
            if row['Destination'] in ['55 Cancri e', 'TRAPPIST-1e']:
                row['HomePlanet'] = 'Europa'
            if row['Destination'] == 'TRAPPIST-1e':
                row['HomePlanet'] = 'Mars'
    return row


def populate_destination(row):
    if pd.isna(row['Destination']):
        if row['HomePlanet'] == 'Europa':
            row['Destination'] = np.random.choice(['55 Cancri e', 'TRAPPIST-1e'])
        elif row['HomePlanet'] == 'Mars':
            row['Destination'] = 'TRAPPIST-1e'
    return row

train_data = train_data.apply(populate_homeplanet, axis=1)
train_data = train_data.apply(populate_destination, axis=1)
```

### Data Transformation

Encoding the categorical data and normalizing the data is remaining now before training the model.

```python
class Transformation:
    def __init__(self, df):
        self.df = df
        self.obj_cols = df.select_dtypes(include=['object']).columns.tolist()
        self.categorical = [col for col in self.obj_cols if df[col].nunique() < 10]
        self.non_cat = [col for col in self.obj_cols if df[col].nunique() >= 10]
        self.scaler = StandardScaler()
        self.encoders = {col: LabelEncoder() for col in self.non_cat}
    
    def encoding(self):
        return pd.get_dummies(self.df, prefix = self.categorical, dtype = "float64")
    
    def label_encode(self):
        for col in self.non_cat:
            self.df[col] = self.encoders[col].fit_transform(self.df[col])
        return self.df
    
    def scaling(self, data):
        return self.scaler.fit_transform(data)
```

### Modeling

```python
# Splitting target and features
X_feat = train_data.drop('Transported', axis = 1)
y_tar = train_data['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_feat,
                                                    y_tar,
                                                    test_size = 0.3,
                                                    random_state = 42)
```

```python
class classificationModel:
    def __init__(self, model, X_train, X_test, y_train, y_test):
        self.X_train = X_train
        self.X_test = X_test
        self.y_train = y_train
        self.y_test = y_test
        self.model = model
        
    def model_training(self):
        self.model.fit(self.X_train, self.y_train)
        y_train_pred = self.model.predict(self.X_train)
        y_test_pred = self.model.predict(self.X_test)
        y_test_prob = self.model.predict_proba(self.X_test)[:, 1]
        
        return y_train_pred, y_test_pred, y_test_prob
        
    
    def model_evaluation(self, train_pred, test_pred, test_prob):
        train_acc = accuracy_score(self.y_train, train_pred)*100
        test_acc = accuracy_score(self.y_test, test_pred)*100
        test_f1 = f1_score(self.y_test, test_pred)
        test_pre = precision_score(self.y_test, test_pred)
        test_recall = recall_score(self.y_test, test_pred)
        test_roc_auc = roc_auc_score(self.y_test, test_prob)
        
        print(f"Model : {self.model}")
        print(f"Accuracy of Training Data: {train_acc}")
        print(f"Accuracy of Testing Data: {test_acc}")
        print(f"F1 Score of Testing Data: {test_f1}")
        print(f"Pecision of Testing Data: {test_pre}")
        print(f"Recall of Testing Data: {test_recall}")
        print(f"ROC AUC of Testing Data: {test_roc_auc}")
        print("\n***CLASSIFICATION REPORT***\n")
        print(classification_report(self.y_test, test_pred))
```
